```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
immune = pd.read_csv("Immune.csv", index_col = 0)
```

```python
scaler = MinMaxScaler()
normalized_df = scaler.fit_transform(immune)
normalized_df = pd.DataFrame(normalized_df, columns=immune.columns, index = immune.index)
```

```python
normalized_df
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(normalized_df,
           cmap="rocket_r",
           linewidths=0.5,
           cbar_kws={"shrink": 0.8, "aspect": 40},
            square=True,
           )
plt.xticks(rotation=45, ha="right", fontsize=14,fontweight="bold", fontfamily="serif");
plt.yticks(fontsize=14,fontweight="bold", fontfamily="serif");
#plt.savefig("Plots/Immune_Cell_Enrichment.png",dpi = 512)
```
