```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("Standarize_Data.csv", index_col=0)
```

```python
data
```

```python
target_file = pd.read_csv("MODIFIED_meta_SRP301424.csv", index_col=0)
target_file.head(5)
```

```python
target = target_file['medulloblastoma subgroup']
target_matrix = target.to_numpy()
target_matrix
```

```python
WNT = pd.read_csv("WNT UP significant.csv")
WNT_s = WNT["Gene"][0:50].to_numpy()
```

```python
SHH = pd.read_csv("SHH UP significant.csv")
SHH_s = SHH["Gene"][0:50].to_numpy()
```

```python
Group3 = pd.read_csv("Group3 UP significant.csv")
Group3_s = Group3["Gene"][0:50].to_numpy()
```

```python
Group4 = pd.read_csv("Group4 UP significant.csv")
Group4_s = Group4["Gene"][0:50].to_numpy()
```

```python
Signnificant  = np.concatenate((WNT_s,SHH_s,Group3_s,Group4_s))
```

```python
Signnificant.shape
```

```python
actual_data = data[Signnificant]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
target_matrix_encoded = le.fit_transform(target_matrix)
target_matrix_encoded
```

```python
X_train, X_test, y_train, y_test = train_test_split(actual_data, target_matrix_encoded,test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
feature_importance = model.coef_[0]
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
from sklearn.preprocessing import label_binarize
y_train_bin = label_binarize(y_train, classes=[0, 1, 2, 3])  # Adjust the classes for your case
y_test_bin = label_binarize(y_test, classes=[0, 1, 2, 3])  # Adjust the classes for your case

# 3. Get the predicted probabilities for each class
y_prob = model.predict_proba(X_test)

# 4. Plot ROC curve for each class
n_classes = y_train_bin.shape[1]
fpr = dict()
tpr = dict()
roc_auc = dict()

plt.figure(figsize=(8, 6))

class_names = ['Group3', 'Group4', 'SHH', 'WNT']

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], lw=2, label=f'{class_names[i]} (AUC = {roc_auc[i]:.2f})')

# 5. Plot the ROC curve for all classes
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')  # Diagonal line (no discrimination)
plt.xlabel('False Positive Rate',fontweight = 'bold',fontfamily="serif")
plt.ylabel('True Positive Rate',fontweight = 'bold',fontfamily="serif")
plt.title('Multiclass ROC Curve (One-vs-Rest)',fontweight = 'bold',fontfamily="serif", fontsize = 18)
plt.legend(loc='lower right')
plt.grid(True)
plt.savefig("ROC_Curve.png", dpi = 512)
plt.show()

# 6. Optionally, compute the average AUC score (macro-average)
macro_auc = auc(fpr[i], tpr[i])  # Macro-average of AUC scores
print(f'Macro-average AUC: {macro_auc:.2f}')
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)

class_labels = ['Group3', 'Group4', 'SHH', 'WNT']


plt.figure(figsize=(8, 6))
sns.heatmap(cm,
            annot=True,
            fmt="d",
            cmap="rocket_r",
            linewidths=0.4,
            linecolor='black',
            square=True,
            xticklabels=class_labels,
            yticklabels=class_labels,)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Group", fontweight = 'bold',fontfamily="serif")
plt.ylabel("Actual Group",fontweight = 'bold',fontfamily="serif")

plt.savefig("Confusion Matrix.png", dpi = 512)
plt.show()
```

```python
report = classification_report(y_test, y_pred,output_dict=True)

report_df = pd.DataFrame(report).transpose()
report_df.iloc[:-3, :-1] *= 100

class_names = ['Group3', 'Group4', 'SHH', 'WNT']
report_df.index = class_names + ['accuracy', 'macro avg', 'weighted avg']

plt.figure(figsize=(10,8))
sns.heatmap(report_df.iloc[:-1, :-1].astype(float),
            annot=True,
            cmap='Blues',
            fmt='.2f',
            linewidths=0.5,
            cbar=True,
           square = True)
plt.title('Classification Report Heatmap',fontweight = 'bold',fontfamily="serif", fontsize = 12)
plt.xlabel('Metrics',fontweight = 'bold',fontfamily="serif")
plt.ylabel('Classes',fontweight = 'bold',fontfamily="serif")
plt.savefig("Classification Report.png", dpi = 512)
plt.show()

```
